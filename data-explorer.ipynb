```python
def plot_current(figsize, x, y):
    fig = plt.figure(figsize=figsize)
    ax = fig.add_subplot(111)
    ax.plot(x,y)
    plt.show()
    
def plot_track_3d(x,y,z):
    ipv.plot(x,y,z)
    ipv.show()
    ipv.clear()
```

```python
i = 1
data = pd.read_csv(os.path.join(final_dir, "%.2d/PLC/plc.csv"%i)) 
plot_track_3d(data.x, data.y, data.z)
plot_current((64,9), range(data.shape[0]), data.spindle_load.abs())
```

```python
def rms_value(array):
    array = array.values.flatten()
    return np.sum(np.square(array))/array.shape[0]
```

```python
feature = group_by_time["spindle_load"].agg([np.sum, np.mean, np.std, rms_value])
feature
```
